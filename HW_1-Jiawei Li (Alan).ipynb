### 5. Write a Python Program to Check Whether a string is palindrome or not

* Initialize a string
* Lower case it to allow easy comparison - Use `my_str.lower()`
* Check using if-else statements if the string is a palindrome
* A palindrome is a string which is same read forward or backwards.
* Alternatively, use the [`extended slice`](https://docs.python.org/3/whatsnew/2.3.html#extended-slices) syntax to reverse a string and check for palindrome

```python
# COMPLETE CELL BELOW

#my_str = 'aIbohPhoBiA' 
# change this value for a different output
def palindrome(my_str):
    my_str=my_str.lower()
    for i in range(len(my_str)):
        if my_str[i]==my_str[-(i+1)]:
            return True
        else:
            return False
            
str_test=input("type a string: ")
print(palindrome(str_test))
        
```

### 7. Write a Python Program to print the first n numbers of the Fibonacci series

* Initialize a variable n that denotes the number of elements of the Fibonacci series to be computed
* Use a loop to print the first n numbers of the Fibonacci series
* Fibonacci series is a recursive series where the `n-th` number in the series is a sum of `n-1`th and `n-2`th element in the series, starting from 0 and 1

```python
# COMPLETE CELL BELOW
def fibonacci(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fibonacci(n-1)+fibonacci(n-2)

# print the first 5 Fibonacci numbers
n = 5
for i in range(n):
    print(fibonacci(i))
```

### 10. Define a class named Triangle that can defines the methods `area` and `perimeter`

* The class can be constructed by the length of the sides. Hint - As arguments of `__init__`
* $area = \sqrt{(s(s-a)(s-b)(s-c)}$ where $s=\frac{(a+b+c)}{2}$ is the semi-perimeter
* $perimeter = a +  b + c$
* **Optional** - If there are repeated calls to the `area` and the `perimeter` functions, how would you optimize the code?

```python
# COMPLETE CELL BELOW
import math

class Triangle:
    def __init__ (self, a, b, c): 
        self.a=a
        self.b=b
        self.c=c
        
    def perimeter(self):
        return self.a+self.b+self.c
    
    def area(self):
        s=(self.a+self.b+self.c)/2
        return math.sqrt(s*(s-self.a)*(s-self.b)*(s-self.c))
    

triangle = Triangle(6,8,10)
                
print("perimeter is:", triangle.perimeter())
print("area is:", triangle.area())


```

### 16. Flip Image using loops

Add a color image of your choosing to the directory. Show the image within the notebook. Extract and show only the red (R of RGB) in one image, only the G in one image, and only B in another image.

```python
from PIL import Image
img = Image.open('nba2k.jpg')
# img.show()
display(img)

def red(picture):
    (width, height)=picture.size
    for x in range(width):
        for y in range(height):
            (r, g, b)=picture.getpixel((x,y))
            picture.putpixel((x,y), (r,0,0))

def green(picture):
    (width, height)=picture.size
    for x in range(width):
        for y in range(height):
            (r, g, b)=picture.getpixel((x,y))
            picture.putpixel((x,y), (0,g,0))
            
def blue(picture):
    (width, height)=picture.size
    for x in range(width):
        for y in range(height):
            (r, g, b)=picture.getpixel((x,y))
            picture.putpixel((x,y), (0,0,b))
red(img)
#img.show()
display(img)

img = Image.open('nba2k.jpg')
green(img)
#img.show()
display(img)

img = Image.open('nba2k.jpg')
blue(img)
#img.show()
display(img)
```

### 17. Flip Image using loops

Use nested for loops to cycle through an image. Create and show an new image flipping the image left to right.

```python
from PIL import Image
img = Image.open('nba2k.jpg')

# use transpose method flip left to right
flip_img = img.transpose(Image.FLIP_LEFT_RIGHT)
display(img)
display(flip_img)

# use loops
img = Image.open('nba2k.jpg')
display(img)

(width, height)=img.size
for y in range(height):
    for x in range(width):
        (r,g,b)=img.getpixel((x,y))
        newx=width-1-x
        flip_img.putpixel((newx,y), (r,g,b))            

display(flip_img)
```

### 18. Filtering

Add your own new image to the directory, create 3 new images using 3 different filters ([Pillow Filter Help](https://pillow.readthedocs.io/en/stable/reference/ImageFilter.html))

```python
from PIL import Image
from PIL import ImageFilter
im = Image.open("nba2k.jpg")
 
# Boxblur radius=2
im1 = im.filter(ImageFilter.BoxBlur(2))
display(im1)

# GaussianBlur radius=2
im2 = im.filter(ImageFilter.GaussianBlur(2))
display(im2)

# Unsharp Mask filter
im3= im.filter(ImageFilter.UnsharpMask(radius=2, percent=200, threshold=5))
display(im3)
```
